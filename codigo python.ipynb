```python
nuevo_orden=['Outcome','Glucose','Insulin','BMI','DiabetesPedigreeFunction','Age']
df = df[nuevo_orden]
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(with_mean=True, with_std=True)

df_std = scaler.fit_transform(df)

# Convertir el array NumPy de nuevo a DataFrame
df_std = pd.DataFrame(df_std, columns=df.columns)


df_std
```

```python
np.mean(df_std[:,0])
```

```python
np.std(df_std[:,0])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

df_pca = pca.fit_transform(df_std)

df_pca = pd.DataFrame(df_pca, columns=["Outcome","Glucose"])

df_pca
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize= (5,5))
plt.rcParams["font.family"] = "serif"

x_label = 'Componente 1 ('+str(round(pca.explained_variance_ratio_[0]*100,2))+'%)'
y_label = 'Componente 2 ('+str(round(pca.explained_variance_ratio_[1]*100,2))+'%)'

ax = fig.add_subplot(1,1,1)
ax.set_xlabel(x_label,fontsize = 10)
ax.set_ylabel(y_label,fontsize = 10)
ax.set_title('Componentes Principales', fontsize = 15, fontstyle = 'italic')
ax.set_xlim(-5,6)

ax.scatter(x = df_pca[:,0], y = df_pca[:,1], s = 20)

plt.show
```
